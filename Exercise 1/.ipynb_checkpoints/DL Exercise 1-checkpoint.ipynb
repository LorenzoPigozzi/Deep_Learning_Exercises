```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('Churn.csv')
```

### Step 1 : 
### Data exploration and preprocessing

```python
# checking the distribution of target labels
df.Exited.value_counts()
```

```python
# splitting in input and output space

# input space
X = df.iloc[:, 3:13].values
# output space
y = df.iloc[:, 13].values
```

```python
# encoding categorical features
ct = ColumnTransformer([('one_hot_encoder', OneHotEncoder(drop="first"), [1,2])],remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
# defining Training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Standardization of the variables
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Step 2 :
### Modeling

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Creating an empty Neural Network
classifier = Sequential()

# adding the first layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
# adding the second layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
# adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

### Step 3 :
### Training process

```python
# defining the optimizer, the loss function and the metric to use
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
# fitting the training set
classifier.fit(X_train, y_train, epochs = 100)
```

### Step 4 :
### Evaluation

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
# displaying the predictionof the target label
y_pred
```

```python
# Creating the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
# dispaying the confusion matrix
cm
```

```python
# Print Statistics from Confusion Matrix
target_names = ['Stay', 'Leave']
print(classification_report(y_test, y_pred, target_names=target_names))
```
